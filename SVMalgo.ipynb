```python
import pandas as pd

path = "C:\\Users\\kavana\\OneDrive\\Desktop\\internship\\Data\\Data\\Social_Network_Ads.csv"

data = pd.read_csv(path)

print(data['Gender'].unique())
print(data['Gender'].nunique())

data['Gender_n'] = data['Gender'].map({'Male':1,'Female':0})
print(data)

inputs = data.drop(['User ID','Purchased','Gender'],'columns')
output = data.drop(['User ID','Gender','Age','EstimatedSalary','Gender_n'],'columns')

print(inputs)
print(output)



import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

x_train,x_test,y_train,y_test = train_test_split(inputs,output,test_size=0.2)

from sklearn.preprocessing import StandardScaler #to increase the accuracy

sc=StandardScaler()

x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)


model = SVC() #initializing the model
model.fit(x_train,y_train) #training

y_pred = model.predict(x_test)
print(y_pred)
print(y_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)

import numpy as np #transforming x_test to more accurate
newinputs=np.array([[45,60000,1]])
newinputs=sc.transform(newinputs)
res=model.predict(newinputs)
print(res)

"""
res=model.predict([[27,57000,0]]) #getting same output for every input because we are not transforming x_test(output will be for 68% accuracy)
print(res)"""
```
