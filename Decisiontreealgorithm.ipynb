```python
import pandas as pd
path="C:\\Users\\kavana\\OneDrive\\Desktop\\internship\\Data\\Data\\salaries.csv"
data=pd.read_csv(path)
print(data)
print(data.info())

import sklearn
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder() #object

data['company_n']=le.fit_transform(data['company'])
data['job_n']=le.fit_transform(data['job'])
data['degree_n']=le.fit_transform(data['degree'])
print(data)


inputs=data.drop(['salary_more_then_100k','company','job','degree'],'columns')
output=data.drop(['company','job','degree','company_n','job_n','degree_n'],'columns')
print(inputs)
print(output)

import sklearn
from sklearn import tree

model=tree.DecisionTreeClassifier()

model.fit(inputs,output)

res=model.predict([[1,2,1]]) #fb,sales execurive,masters
print(res)

res=model.predict([[2,2,1]]) #google,SE,masters
print(res)

res=model.predict([[0,0,1]]) #abcpharma,BM,masters
print(res)

res=model.predict([[1,1,1]]) #fb,CP,masters
print(res)

```
